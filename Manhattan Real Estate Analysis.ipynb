<h3><b>Imputation of the columns SALE PRICE, YEAR BUILT, TOTAL UNITS, TAX CLASS AT PRESENT and BUILDING CLASS AT PRESENT. Values with 0, NaN and that are missing are replaced with the mean, median and mode values of that column </b></h3>

```python
# Imputation of columns SALE PRICE and YEAR BUILT. Ancestarel inherited property values which are having SALE PRICE 0,
# are imputed with value 1 and similarly the columns which are YEAR BUILT as 0 is imputed with the mean value of the YEAR BUILT. 
df_v_yb = df
df_v_yb['SALE PRICE'] = df_v_yb['SALE PRICE'].apply(lambda x: 1 if x == 0 else x)
mean_yearbuilt = df_v_yb['YEAR BUILT'].mean().astype('int')
median_totalunits = df_v_yb['TOTAL UNITS'].median().astype('int')
df_v_yb['YEAR BUILT'] = df_v_yb['YEAR BUILT'].apply(lambda x: mean_yearbuilt if x == 0 else x)
dummy_df = df_copy2
df_v_yb['TOTAL UNITS'] = dummy_df['TOTAL UNITS']
df_v_yb['TOTAL UNITS'] = df_v_yb['TOTAL UNITS'].fillna(df_v_yb['TOTAL UNITS'].median())
df_v_yb['TOTAL UNITS'] = df_v_yb['TOTAL UNITS'].astype('int')
df_v_yb['TOTAL UNITS']
tax_mode = df_v_yb["TAX CLASS AT PRESENT"].mode()[0]
df_v_yb["TAX CLASS AT PRESENT"] = df_v_yb["TAX CLASS AT PRESENT"].fillna(tax_mode)
bldg_mode = df_v_yb["BUILDING CLASS AT PRESENT"].mode()[0]
df_v_yb["BUILDING CLASS AT PRESENT"] = df_v_yb["BUILDING CLASS AT PRESENT"].fillna(bldg_mode)
print(df_v_yb.isna().sum())
```
